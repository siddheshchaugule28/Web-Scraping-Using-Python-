```python
# Importing necessary libraries

import pandas as pd            # For data manipulation and analysis
import requests                # For sending HTTP requests to the website
from bs4 import BeautifulSoup  # For parsing HTML content
```

```python
# Setting up the URL for web scraping

url="https://www.flipkart.com/search?q=laptops&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_5_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_5_0_na_na_na&as-pos=5&as-type=TRENDING&suggestionId=laptops&requestId=0848e20d-7c82-40f3-a041-d906b606c419"
```

```python
# Defining user-agent and language preferences for the request

HEADERS=({'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36","Accept-Language":"en-US, en;q=0.5"})
```

```python
# Sending a GET request to the URL

r=requests.get(url,headers=HEADERS)
print(r)   # Print the response status (200 if successful)
```

```python
# Parsing the HTML content of the page using BeautifulSoup

soup=BeautifulSoup(r.text,"lxml")
```

```python
# Print the prettified HTML content for better readability

print(soup.prettify())
```

```python
# Lists to store scraped data
Product = []        # To store product names
features = []       # To store product features
Price = []          # To store product prices
Rating = []         # To store product ratings
Total_reviews = []  # To store total reviews of the product


# Function to scrape laptop details from the page

def laptop_scrap():
    for device in soup.find_all(class_="_3pLy-c row"):
        try: 
            Product.append(device.find(class_="_4rR01T").text.strip())
        except:
            Product.append(None)
        try:
            features.append(device.find(class_="_1xgFaf").text.strip())
        except:
            features.append(None)
        try: 
            Price.append(device.find(class_="_30jeq3 _1_WHN1").text.strip())
        except:
            Price.append(None)
        try:
            Rating.append(device.find(class_="_3LWZlK").text.strip())
        except:
            Rating.append(None)
        try:
            Total_reviews.append(device.find(class_="_2_R_DZ").text.strip())
        except:
            Total_reviews.append(None)
```

```python
# Creating DataFrame to store the scraped data            

df1=pd.DataFrame({"Product":Product,
                  "features":features,
                  "Price":Price,
                  "Rating":Rating,
                  "Total_reviews":Total_reviews,
                 })
```

```python
# Total number of pages to scrape

total_pages=81

# Looping through each page to scrape data

for page in range(total_pages):
    # Calculate progress percentage
    progress = (page + 1) / total_pages * 100
    # Display progress and page number
    print(f"Progress: {progress:.2f}%", f"Page No.:{page+1}", end="\r")
    # Construct URL for the next page
    url = f"https://www.flipkart.com/search?q=laptops&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_5_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_5_0_na_na_na&as-pos=5&as-type=TRENDING&suggestionId=laptops&requestId=0848e20d-7c82-40f3-a041-d906b606c419&page={page}"
    # Send request to the next page and parse HTML
    request = requests.get(url).text
    soup = BeautifulSoup(request, "lxml")
    # Call function to scrape laptop details from the current page
    laptop_scrap()
```

```python
# Creating DataFrame to store all scraped data

df2=pd.DataFrame({"Product":Product,
                  "features":features,
                  "Price":Price,
                  "Rating":Rating,
                  "Total_reviews":Total_reviews,
                 })
```
